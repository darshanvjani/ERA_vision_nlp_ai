```python
from Utilities.config import get_config

__conf__ = get_config()
__conf__['batch_size'] = 16
__conf__['preload'] = None
__conf__['num_epochs'] = 30

from Utilities.dataset import BilingualDataModule
from Utilities.model import build_transformer
import pytorch_lightning as pl
from pytorch_lightning.callbacks import ModelCheckpoint, LearningRateMonitor
```

```python
datamodule = BilingualDataModule(__conf__)
datamodule.setup()
```

```python
model = build_transformer(
    datamodule.tokenizer_src.get_vocab_size(),
    datamodule.tokenizer_tgt.get_vocab_size(),
    __conf__["seq_len"],
    __conf__["seq_len"],
    d_model=__conf__["d_model"],
)
```

```python
trainer = pl.Trainer(
    max_epochs=__conf__["num_epochs"],
    accelerator="gpu",
    callbacks=[
        ModelCheckpoint(dirpath=__conf__['model_folder'], verbose=True,), 
        LearningRateMonitor(logging_interval='step', log_momentum=True)
    ],
    default_root_dir="Store/",
    fast_dev_run=False,
    check_val_every_n_epoch=2,
    num_sanity_val_steps=10,
    limit_val_batches=100,
)
```

```python
trainer.fit(model, datamodule)
```
